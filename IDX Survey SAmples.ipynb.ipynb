```
# Step 1: Import necessary libraries
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt
import seaborn as sns




df = pd.read_excel('/content/IDC_Survey_Sample.xlsx')

df.head()
```

```
print(df.isnull().sum())
```

```
# Count columns before cleaning
num_columns_before = df.shape[1]
print(f"Number of columns before cleaning: {num_columns_before}")

# Clean columns with null values
df_cleaned = df.dropna(axis=1)

# Count columns after cleaning
num_columns_after = df_cleaned.shape[1]
print(f"Number of columns after cleaning: {num_columns_after}")

# Display the cleaned DataFrame
df_cleaned.head()
```

```
# Filter for 'Successful' entries in the 'Verbatim' column
successful_calls = df[df['Verbatim'] == 'Successful']

# Count successful calls by agent
successful_call_counts = successful_calls['Agent Name'].value_counts().reset_index()
successful_call_counts.columns = ['Agent Name', 'Successful Count']

# Plotting
plt.figure(figsize=(1, 6))  # Adjust the figure size as needed
sns.barplot(x='Agent Name', y='Successful Count', data=successful_call_counts, palette='Greens_d')
plt.title('Successful Calls by Agent Name')
plt.xlabel('Agent Name')
plt.ylabel('Number of Successful Calls')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```
