```python
import pandas as pd
import plotly.express as px

# Load the data
senora_survey = pd.read_excel('https://github.com/sadathasan4/cdsrtextanalytics/raw/refs/heads/main/Senora_Survey_Sample1.xlsx')

# Extract numeric family income
def extract_numeric_income(income):
    if pd.isna(income):  # Handle NaN values
        return None
    income = income.replace(' taka', '')  # Remove 'taka'
    if 'Above' in income:
        return int(income.replace('Above ', '').strip())  # Return the numeric value above a threshold
    elif 'Below' in income:
        return int(income.replace('Below ', '').strip())  # Return the numeric value below a threshold
    elif '-' in income:
        low, high = map(int, income.split('-'))  # Split and convert to int
        return (low + high) / 2  # Return the average of the range
    else:
        return None  # Handle any unexpected formats

# Apply the function to the Family Income column
senora_survey['Numeric Family Income'] = senora_survey['Family Income'].apply(extract_numeric_income)

# Calculate the average family income by Address and Profession
average_income = senora_survey.groupby(['Address', 'Profession'])['Numeric Family Income'].mean().reset_index()

# Create a bar chart
fig = px.bar(average_income, x='Address', y='Numeric Family Income', color='Profession',
             title='Average Family Income by Address and Profession',
             labels={'Numeric Family Income': 'Average Family Income (Currency)', 'Address': 'Address'},
             barmode='group')
fig.show()
```

```python
# Create a box plot
fig2 = px.box(senora_survey, x='Profession', y='Numeric Family Income',
               title='Distribution of Family Income by Profession',
               labels={'Numeric Family Income': 'Family Income (Currency)', 'Profession': 'Profession'})
fig2.show()
```

```python
# Calculate total family income by profession
total_income_by_profession = senora_survey.groupby('Profession')['Numeric Family Income'].sum().reset_index()

# Create a pie chart
fig6 = px.pie(total_income_by_profession, names='Profession', values='Numeric Family Income',
               title='Total Family Income Distribution by Profession')
fig6.show()
```
